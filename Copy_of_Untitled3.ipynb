```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("/content/car_sales_data.csv")

dataset.head(10)
```

```python
from datetime import datetime
pd.to_datetime(dataset['Car Year'])
```

```python
dataset['Sale Price']=dataset['Sale Price'].astype(float)
```

```python
dataset.describe()
```

```python
dataset.nunique()
```

```python
dataset.duplicated().sum()
```

```python
dataset=dataset.dropna()
```

```python
dataset.isnull().sum()
```

```python
dataset['Car Year']=pd.DatetimeIndex(dataset['Car Year']).year

dataset['Car Year'].value_counts()
```

```python
dataset.columns
```

```python
import matplotlib.pyplot as plt

model_counts = dataset['Car Model'].value_counts()

plt.plot(model_counts.index, model_counts.values)
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.title('Count of Car Model')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(dataset['Commission Rate'], bins=2)
plt.xlabel('Commision Rate')
plt.ylabel('rate')
plt.title('Distribution of commision Rate')
plt.show()
```

```python
model_counts=dataset['Car Model'].value_counts()

plt.pie(model_counts, labels=model_counts.index, autopct='%1.0f%%',startangle=90)
plt.title('Distribution of Car Model')
plt.show()
```

```python
import matplotlib.pyplot as plt

model_counts = dataset['Car Model'].value_counts()

plt.scatter(model_counts.index, model_counts.values)
plt.xlabel('Model')
plt.ylabel('Count')
plt.title('Count of Items in Each model')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

make_counts = dataset['Car Make'].value_counts()

plt.bar(make_counts.index, make_counts.values)
plt.xlabel('Car Make')
plt.ylabel('Count')
plt.title('Count of Car Make')
plt.xticks(rotation=45)
plt.show()
```
